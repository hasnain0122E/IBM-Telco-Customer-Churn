### OBJECTIVE

Convert the feature-engineered dataset into a format that machine learning algorithms can understand while following industry best practices.

```python
import pandas as pd 
import numpy as np 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import ( OneHotEncoder, StandardScaler )

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("../data/featured/telco_feature_engineered.csv")
```

```python
x = df.drop(columns = 'Churn')
y = df['Churn']
```

```python
x.columns.tolist()
```

```python
drop_cols = ['customerID']
x = x.drop(columns= drop_cols)
```

```python
X_train, X_test, y_train, y_test  = train_test_split(
    x,
    y,
    test_size=0.2, # 20% training data splitted
    random_state = 42,
    stratify=y
)
```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
print("Original:")
print(y.value_counts(normalize=True) * 100)

print("\nTraining:")
print(y_train.value_counts(normalize=True) * 100)

print("\nTesting:")
print(y_test.value_counts(normalize=True) * 100)
```

### Identify Feature Types

```python
X_test.dtypes
```

```python
numerical_features = ['SeniorCitizen', 'tenure','MonthlyCharges', 'TotalCharges', 'TotalServices','HasSecurityBundle','AverageMonthlySpend']
```

```python
categorical_features = ["gender",
    "Partner",
    "Dependents",
    "PhoneService",
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod",
    "TenureGroup",
    "IsFamilyCustomer"]
```

```python
X_train.columns.tolist()
```

### Create the Transformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import ( StandardScaler, OneHotEncoder)

preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            StandardScaler(),
            numerical_features
        ),
        (
            "cat",
            OneHotEncoder( handle_unknown="ignore"),
            categorical_features
        )
    ]
)
```

### Fit and Transform

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
print(type(X_train_processed))
print(X_train_processed.shape)
print(X_test_processed.shape)
```

```python
feature_names = preprocessor.get_feature_names_out()
print(len(feature_names))
```

```python
print(feature_names[:20])
```

### First Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
pipeline = Pipeline(
    steps=[
        ("preprocessor",  preprocessor),
        ("Classifier", LogisticRegression(random_state=42, max_iter=1000))
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
prediction = pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, prediction)
print("Accuracy", accuracy*100)
```
